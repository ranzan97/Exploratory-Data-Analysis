```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load the dataset
url = "https://raw.githubusercontent.com/nikhilsthorat03/Telco-Customer-Churn/refs/heads/main/telco.csv"
data = pd.read_csv(url)
```

```python
# Step 2: Basic Information
print("Dataset Shape:", data.shape)
print("\nFirst 5 Rows of the Dataset:")
print(data.head())
```

```python
print("\nDataset Info:")
print(data.info())
```

```python
print("\nSummary Statistics:")
print(data.describe(include='all'))
```

```python
# Step 3: Check for Missing Values
print("\nMissing Values in Each Column:")
print(data.isnull().sum())
```

```python
# Step 4: Distribution of Target Variable ('Churn')
plt.figure(figsize=(8, 6))
sns.countplot(x='Churn', data=data, palette='Set2')
plt.title('Distribution of Churn')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.show()
```

```python
# Step 5: Numerical Features Analysis
numerical_cols = data.select_dtypes(include=[np.number]).columns
```

```python
# Correlation Heatmap for Numerical Features
plt.figure(figsize=(12, 8))
sns.heatmap(data[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap for Numerical Features')
plt.show()
```

```python
# Distribution of Numerical Features
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[col], kde=True, bins=30, color='blue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Step 6: Categorical Features Analysis
categorical_cols = data.select_dtypes(include=['object']).columns
```

```python
# Countplots for Categorical Features
for col in categorical_cols:
    if col != 'customerID':  # Skip customerID as it's unique
        plt.figure(figsize=(8, 4))
        sns.countplot(x=col, data=data, palette='Set2', order=data[col].value_counts().index)
        plt.title(f'Distribution of {col}')
        plt.xticks(rotation=45)
        plt.show()
```

```python
# Step 7: Relationship Between Features and Target Variable ('Churn')
for col in categorical_cols:
    if col not in ['customerID', 'Churn']:
        plt.figure(figsize=(8, 4))
        sns.countplot(x=col, hue='Churn', data=data, palette='Set2')
        plt.title(f'{col} vs Churn')
        plt.xticks(rotation=45)
        plt.show()
```

```python
# Step 8: Pairplot for Numerical Features
sns.pairplot(data[numerical_cols])
plt.show()
```

```python
# Step 9: Boxplots for Numerical Features vs Target Variable ('Churn')
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='Churn', y=col, data=data, palette='Set2')
    plt.title(f'{col} vs Churn')
    plt.show()
```
