```python
# 1. Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Heart Dataset
url = 'https://raw.githubusercontent.com/sharmaroshan/Heart-UCI-Dataset/refs/heads/master/heart.csv'
heart = pd.read_csv(url)
```

```python
heart.head()
```

```python
heart.isnull().sum()
```

```python
heart.describe()
```

```python
# Distribution of target variable (Heart Disease vs No Heart Disease)
sns.countplot(x='target', data=heart)
plt.title('Distribution of Heart Disease (0 = No, 1 = Yes)')
plt.show()
```

```python
# Age distribution
sns.histplot(heart['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Heart Disease by Gender
sns.countplot(x='sex', hue='target', data=heart)
plt.title('Heart Disease by Gender (0 = Female, 1 = Male)')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heart.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter plot of Age vs Max Heart Rate (thalach) with hue for target
sns.scatterplot(x='age', y='thalach', hue='target', data=heart)
plt.title('Age vs Max Heart Rate (Thalach) by Heart Disease')
plt.show()
```

```python
# Boxplot of Cholesterol levels by Heart Disease
sns.boxplot(x='target', y='chol', data=heart)
plt.title('Cholesterol Levels by Heart Disease')
plt.show()
```

```python
# Pairplot to visualize relationships between key numerical variables
sns.pairplot(heart[['age', 'trestbps', 'chol', 'thalach', 'target']], hue='target')
plt.show()
```

```python
print("\nCorrelation matrix:")
print(heart.corr())
```
